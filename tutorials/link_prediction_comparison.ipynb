# Подсчитаем точность восстановления ребер на различных loss functions и датасатах

```python
import torch_geometric.transforms as T
from torch_geometric.datasets import Planetoid
from stable_gnn.model_link_predict import ModelLinkPrediction
```

# Первый датасет - CORA

```python
root = "../tmp/"
name = "Cora"
dataset = Planetoid(root=root + str(name), name=name, transform=T.NormalizeFeatures())
```

```python
loss_functions = ['APP', 'LINE', 'HOPE_AA', 'VERSE_Adj']
for loss in loss_functions:
    #if loss is None:
    #annotation = 'Train on attributes, without loss functions'
    #model = ModelLinkPrediction(number_of_trials=0)
    #print(annotation)
   # else:
    #    annotation=''
    model = ModelLinkPrediction(number_of_trials=50,loss_name=loss)

    train_edges, train_negative, test_edges, test_negative = model.train_test_edges(dataset)
    cl = model.train_cl(train_edges,train_negative)
    #print(annotation)
    print("For loss: {}, f-score is {}".format(loss,model.test(cl,test_edges,test_negative)))
```
